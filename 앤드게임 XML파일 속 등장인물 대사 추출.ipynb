```python
import re
import xml.etree.ElementTree as ET

# Define the file paths
xml_file_path = '/content/Avengers_Endgame_xml.xml'
txt_file_path = '/content/Tony_Speeches.txt'

# Read the XML file as a string and remove any problematic characters
with open(xml_file_path, 'r', encoding='utf-8') as file:
    xml_content = file.read()

# Optional: Clean the XML content if there are any specific problematic characters or issues
xml_content = re.sub(r'[^\x09\x0A\x0D\x20-\x7F]+', '', xml_content)

# Parse the XML content from the cleaned string
root = ET.fromstring(xml_content)

# Open the text file in write mode
with open(txt_file_path, 'w') as txt_file:
    # Iterate through the XML tree to find all SPEECH tags
    for speech in root.findall('.//SPEECH'):
        speaker = speech.find('SPEAKER')
        if speaker is not None and speaker.text == 'TONY':
            # Extract and write the speech lines to the text file
            lines = speech.findall('LINE')
            for line in lines:
                if line.text:  # Ensure the line is not None
                    txt_file.write(line.text.strip() + '\n')
            txt_file.write('\n')  # Add a newline for separation between speeches

print(f"Extracted speeches saved to: {txt_file_path}")
```

```python
from bs4 import BeautifulSoup

# Define the file paths
xml_file_path = '/content/Avengers_Endgame_xml.xml'
txt_file_path = '/content/Tony_Speeches.txt'

# Read the XML file as a string
with open(xml_file_path, 'r', encoding='utf-8') as file:
    xml_content = file.read()

# Parse the XML content using BeautifulSoup
soup = BeautifulSoup(xml_content, 'xml')

# Open the text file in write mode
with open(txt_file_path, 'w') as txt_file:
    # Find all SPEECH tags
    speeches = soup.find_all('SPEECH')
    for speech in speeches:
        speaker = speech.find('SPEAKER')
        if speaker and speaker.text.strip() == 'TONY':
            # Extract and write the speech lines to the text file
            lines = speech.find_all('LINE')
            for line in lines:
                if line.text:  # Ensure the line is not None
                    txt_file.write(line.text.strip() + '\n')
            txt_file.write('\n')  # Add a newline for separation between speeches

print(f"Extracted speeches saved to: {txt_file_path}")
```
