```python
import xml.etree.ElementTree as ET

def convert_to_xml(text):
    root = ET.Element("movie")
    current_scene = None
    last_speaker = None

    for line in text.split('\n'):
        line = line.strip()
        if line.startswith('[') and line.endswith(']'):
            if current_scene is not None:
                root.append(current_scene)
            current_scene = ET.Element("scene")
            description = ET.SubElement(current_scene, "description")
            description.text = line
        elif ': ' in line:
            speaker, dialogue = line.split(': ', 1)
            if current_scene is not None:
                dialogue_elem = ET.SubElement(current_scene, "dialogue")
                speaker_elem = ET.SubElement(dialogue_elem, "speaker")
                speaker_elem.text = speaker
                line_elem = ET.SubElement(dialogue_elem, "line")
                line_elem.text = dialogue
                if last_speaker:
                    receiver_elem = ET.SubElement(dialogue_elem, "receiver")
                    receiver_elem.text = last_speaker
                last_speaker = speaker
        elif current_scene is not None:
            description = ET.SubElement(current_scene, "description")
            description.text = line

    if current_scene is not None:
        root.append(current_scene)

    tree = ET.ElementTree(root)
    return tree

def main():
    # 입력 파일 경로 설정
    input_file_path = "/content/infinity_war.txt"

    # 출력 파일 경로 설정 (여기에 XML 파일을 저장할 경로를 입력하세요)
    output_file_path = '/content/output.xml'

    # 텍스트 파일 읽기
    with open(input_file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    # XML로 변환
    tree = convert_to_xml(text)

    # XML 파일로 저장
    tree.write(output_file_path, encoding='utf-8', xml_declaration=True)
    print(f"XML 파일이 {output_file_path}에 저장되었습니다.")

if __name__ == "__main__":
    main()

```
